## 2. Data Preprocessing and Cleaning

```python
# drop datetime info
# -----------
chicago_inspections_2011_to_2013 = chicago_inspections_2011_to_2013.dropna().drop(['Inspection_Date',
                                                                                   'minDate',
                                                                                   'maxDate',
                                                                                   'Inspection_MonthYear'],
                                                                                  axis = 1)
```

```python
# process target 
# -----------
y = chicago_inspections_2011_to_2013['Results']

# decide if you want to binarize the outcome variable 
# -----------
# comment out the following lines of code if you don't want to binarize the target variable
y = y.replace({'Pass w/ Conditions': 'Pass'})
lb_style = LabelBinarizer()
y = lb_style.fit_transform(y)

# recode 0s and 1s so 1s are "Fail"
y = np.where(y == 1, 0 ,1)


# process features
# -----------

# create feature dataset
X = chicago_inspections_2011_to_2013.drop(columns = ['Results', 
                                                     'Facility_Type', 
                                                     'LICENSE_DESCRIPTION', 
                                                     'WARD_PRECINCT'])


# get dummies
X = pd.get_dummies(X)
```

```python
# data splitting
# import train test split
from sklearn.model_selection import train_test_split

# random seed
np.random.seed(42)

# split into train and validation sets
X_train, X_val, y_train, y_val = train_test_split(
    X, y,
    test_size=0.2,
    stratify=y,
    random_state=42
)

# check split
print(f"Training set size: {X_train.shape}")
print(f"Validation set size: {X_val.shape}")
print(f"Training failures: {y_train.sum()}")
print(f"Validation failures: {y_val.sum()}")
```

### Model 1: Logistic Regression

Logistic regression is both a classification and regression method (James et al., 2023). Logistic regression provides a combination of variables and suggests relative variable importance (Breiman, 2001). Here logistic regression models probability of inspection failure. An advantage is its interpretability, though its sensitive to outliers and thus may underfit complex patterns.

Here, logistic regression is appropriate because food inspection outcomes likely have some interpretable linear relationships (e.g. more past violations => higher failure probability). The model's interpretability would be valuable for explaining decisions to policymakers and restaurant owners without a quantitative research backgroun.


```python
# import library from sklearn
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import GridSearchCV

# logistic regression setup
log_reg = LogisticRegression(random_state=42, max_iter=1000)

# make hyperparameter grid
param_grid = {
    'C' : [0.01, 0.1, 1, 10, 100],
    'penalty' : ['l1', 'l2'],
    'solver' : ['liblinear']
}

# grid search
grid_lr = GridSearchCV(
    log_reg,
    param_grid,
    cv=5,
    scoring='f1',
    n_jobs=-1,
    verbose=1
)

# fit model
grid_lr.fit(X_train, y_train)

# get best model
best_lr = grid_lr.best_estimator_

print(f"Best parameters: {grid_lr.best_params_}")
print(f"Best CV score: {grid_lr.best_score_:.3f}")
```

###  Model 2: XG Boost

XG Boost builds trees in sequence. Each new tree corrects previous trees' errors. Regularization prevents overfitting. XGBoost handles missing values automatically. Hyperparameter tuning can be tedious, but there are worse things in life. For explanation to stakeholders, XGBoost can be more complex than logistic regression.

Food inspection data has class imbalance, complex/non-linear patterns, and XG Boost performs well in these conditions.

### Model 3: Random Forest

Random Forests do a lot of random sampling, including from subsets of predictors, observationsthen, RF trims the sample into leaves (Lundberg et al., 2022). Leaves make trees, and the trees averaged are a forest (which does not make sense, but I appreciate the poetic inclinations). RFs sampling logic tries to prevent overfitting (Reades et al., 2019). However, RFs may be harder to interpret than logistic regression.

An inspection failure could result from many different factors. Random Forests can capture these patterns while remaining robust.


```python
from sklearn.ensemble import RandomForestClassifier

# set up RF
rf = RandomForestClassifier(random_state=42, n_jobs=-1)

# define hyperparameter grid
param_grid_rf = {
    'n_estimators' : [100, 200, 300],
    'max_depth' : [10, 20, 30, None],
    'min_samples_split' : [2, 5, 10],
    'min_samples_leaf' : [1, 2, 4],
    'max_features' : ['sqrt', 'log2']
}

# grid search
from sklearn.model_selection import RandomizedSearchCV

grid_rf = RandomizedSearchCV(
    rf,
    param_grid_rf,
    n_iter=20,
    cv=5,
    scoring='f1',
    n_jobs=-1,
    verbose=1,
    random_state=42
)

# fit model
grid_rf.fit(X_train, y_train)

# best model
best_rf = grid_rf.best_estimator_

print(f"Best parameters: {grid_rf.best_params_}")
print(f"Best CV score: {grid_rf.best_score_:.3f}")
```

## 5. Discussion Questions

1. Why do we need metrics beyond accuracy when using machine learning in the social sciences and public policy?


Accuracy is dangerously misleading when classes are imbalanceda common situation in social science and policy applications. As Prof. Zapatka noted, "accuracy is an expression of ratio of correct observations relative to incorrect observations. This calculation alone does not tell us much about whether we did a good job predicting all of the other categories that we might be concerned about" (class, 2025). In the food inspection data approximately 75% of establishments pass inspections. A nave model that always predicts "Pass" would achieve 75% accuracy while being completely useless. That is to say it would miss 100% of violations!! Accuracy solo is insufficient for evaluating model performance in real-world applications.

Multiple metrics matter because different types of errors carry different consequences. False negatives could cause foodborne illness outbreaks, hospitalizations, deaths, etc, while false positives waste inspection time which I suppose is at least a minor business inconvenience. Accuracy treats these errors equall.


2. Imagine that establishments learned about the algorithm being used to determine who gets audited and they started adjusting their behavior (and changing certain key features about themselves that were important for the prediction) to avoid detection. How could policymakers address this interplay between algorithmic decisionmaking and real world behavior?

If establishments learn about the algorithm, a risk is they may rationally adjust behavior to avoid detection. Establishments might engage in strategies including feature manipulation, superficial, or strategic timing for high-errisk activities. This would all be at the detriment of food safety.

Policymakers can counter malintentioned strategies through counterstrategies of their own. They could pursue opacity and unpredictability by not publishing exact feature weights or by adjusting algorithms , rotating algorithms periodically, and using ensemble methods that are harder to reverse-engineerthough this must be balanced against transparency needed for accountability. They could also (pay researcher to) include a random component. I believe that retraining models frequently with new data would also be helpful.

Ultimately though it doesnt matter why the establishments change their practices (like, if theyre changing just to game the system for the algorithm), as long as the changes align with the policymakers goal of food safety (if that even is the policymakers goal, but I feel like if I start writing about that I will rant).
